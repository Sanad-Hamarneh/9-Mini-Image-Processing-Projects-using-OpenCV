```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import imutils
import random
```

```python
img = cv2.imread("someshapes.jpg")
plt.imshow(img)
```

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
```

```python
edges_can =cv2.Canny(grey_img,250,0)
edges_can = cv2.dilate(edges_can,None)
plt.imshow(imutils.opencv2matplotlib(edges_can))
plt.show()
```

```python
contours = cv2.findContours(image=edges_can, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)[0]

image_copy = img.copy()
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(imutils.opencv2matplotlib(image_copy))
plt.show()
```

```python
cont = sorted(cv2.findContours(edges_can, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0], key=cv2.contourArea)
```

```python
mask1 = np.zeros(grey_img.shape, np.uint8)
mask1.fill(255)
square_b=[cont[0]]
square = cv2.drawContours(mask1,square_b,-1, 0,-1)
```

```python
mask2 = np.zeros(grey_img.shape, np.uint8)
mask2.fill(255)
traingle_b=[cont[2]]
traingle = cv2.drawContours(mask2,traingle_b,-1, 0, -1)
```

```python
mask3 = np.zeros(grey_img.shape, np.uint8)
mask3.fill(255)
star_b=[cont[3]]
star= cv2.drawContours(mask3,star_b,-1, 0, -1)
```

```python
mask4 = np.zeros(grey_img.shape, np.uint8)
mask4.fill(255)
rectangle_b=[cont[6]]
rectangle = cv2.drawContours(mask4,rectangle_b,-1, 0, -1)
```

```python
mask5 = np.zeros(grey_img.shape, np.uint8)
mask5.fill(255)
circle_b=[cont[8]]
circle = cv2.drawContours(mask5,circle_b,-1, 0, -1)
```

```python
images= [square, traingle, star, rectangle , circle]

random.shuffle(images)
shuffled_images = []

for x in images:
    shuffled_images.append(x)

shuffled_images = cv2.hconcat(shuffled_images)

plt.imshow(cv2.cvtColor(shuffled_images, cv2.COLOR_BGR2RGB))
plt.show()
```
